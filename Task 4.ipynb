```python
# changing the names of columns because the dataset given to us haslots of unusable columns and 
# name also not written in proper condition.
df.rename(columns = {'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
#new dataset form folling given valriable which is usefull tp us
df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive','ransompaid','ransomamt']]
```

```python
df.head()
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'Region', data = df, palette = 'rocket_r', order = df['Region'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

#### Middle East  & North Africa, South Asia region has highest terror attack

#### Astralasia & oceania has least terror attack

```python
plt.figure(figsize = (10,12))
sns.countplot(x = "Attacktype", data = df, order = df['Attacktype'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

#### Most of the terror attacks are done by Bombarding and Explosion.
#### The least terror attack are from Hijacking.

```python
plt.figure(figsize = (10,12))
sns.countplot(x = 'Year', data = df, palette = 'rainbow', order = df['Year'].value_counts().index)
plt.title('Attack per Year')
plt.ylabel('No of Attacks')
plt.xticks(rotation = 90)
plt.show()
```

#### In Year 2014, most of the error attacks has been occured.

```python
plt.figure(figsize = (10,5))
country_kill = df.groupby('Year')['kill'].sum()
top_10_country = country_kill.sort_values(ascending = False).iloc[0:10]
top_10_country.plot(kind= 'barh')
plt.title('People Died due to Attack')
plt.show()
```

```python
total_ransom = df.groupby('Country')['ransomamt'].sum()
top_10_country = total_ransom.sort_values(ascending = False).iloc[0:10]
top_10_country.plot(kind = 'barh')
plt.title('Ransome country')
plt.show()
```

#### Italy has the highest ransome than other countries. 

```python
plt.figure(figsize = (10,12))
sns.barplot(x = df['Country'].value_counts()[:10].index, y = df['Country'].value_counts()[:10].values, palette ='rocket_r')
plt.title('Affected countries')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

#### Most of the attacks happened in Iraq,Pakistan,Afganistan and India.

```python
plt.subplots(figsize = (10,12))
sns.barplot(x = df['City'].value_counts()[:10].index, y = df['City'].value_counts()[:10].values, palette = 'rainbow')
plt.title('Affected Cities')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.xticks(rotation =50)
plt.show()
```

#### Most of the attack occures in unknown cities

```python
plt.figure(figsize = (20,10))
sns.barplot(x = df['Group'].value_counts()[:12].index, y = df['Group'].value_counts()[:12].values, palette ='cividis')
plt.title('Groups behind all attacks')
plt.xlabel('Group')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

#### Most of the attacks are done by unknown groups.
#### Taliban, (SI), (ISIL) are activated in large numbers.

```python
plt.figure(figsize = (10,5))
sns.barplot(x = df['Targettype'].value_counts()[:10].index, y = df['Targettype'].value_counts()[:10].values, palette = 'rocket_r')
plt.title('Targetted People')
plt.xlabel('Targets')
plt.ylabel('count')
plt.xticks(rotation = 90)
plt.show()
```

#### mostly Private citizens & Property  and Military are targetted.
